```python
class student:

  # In python there two main type of variable in class
  # 1-> instance variable ( belongs to each object separately)
  # 2-> class variable (static) ( belongs to all object, same for all)

  CLASS_YEAR = 2023 # This is class variable (same for all object)


  def __init__(self):
    self.roll = int(input("Enter your roll no: ")) # self.roll and self.name are instance variable
    self.name = input("Enter your name: ")

  def get_data(self):
    print(f"Roll No: {self.roll}\nName: {self.name}")

pranav = student()
pranav.get_data()
print(pranav.CLASS_YEAR)
```
