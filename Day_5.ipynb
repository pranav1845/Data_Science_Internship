```python
class student:

  # In python there two main type of variable in class
  # 1-> instance variable ( belongs to each object separately)
  # 2-> class variable (static) ( belongs to all object, same for all)

  CLASS_YEAR = 2023 # This is class variable (same for all object)


  def __init__(self):
    self.roll = int(input("Enter your roll no: ")) # self.roll and self.name are instance variable
    self.name = input("Enter your name: ")

  def get_data(self):
    print(f"Roll No: {self.roll}\nName: {self.name}")

pranav = student()
pranav.get_data()
print(pranav.CLASS_YEAR)
```

```python
# Single inheritance

class parent:
  def do_work(self):
    print("parent")

class child(parent):
  def child_work(self):
    print("child")

Nick = child()
Nick.child_work()
Nick.do_work()
```

```python
# multiple inheritance

class university:
  def uni(self):
    print("this is university method")

class college:
  def col(self):
    print("this is college method")

class dept(university, college):
  def dep(self):
    print("hello student")

comp = dept()

comp.col()
```

```python
# multi-level inheritance

class human:

  def drink(self):
    print("human drinking")

  def eat(self):
    print("human eating")

class richhuman(human):

  def balance(self):
    print("rich human")

class Lohar(richhuman):

  def get(self, richhuman):
    print("you are: ")
    richhuman.balance()

Yash = Lohar()
dummy = richhuman()

Yash.drink()
Yash.eat()
Yash.balance()
Yash.get(dummy)
```

```python
#

class Rectangle:
  def __init__(self, x, y) -> None:
    self.x = x
    self.y = y

  def area(self):
    return self.x * self.y


class Square(Rectangle):

  def __init__(self, side) -> None:
    super().__init__(side, side)

redbox = Square(50)

print(redbox.area())
```
