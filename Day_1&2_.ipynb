```python
# Task : Create Simple Calculator in Python using Appropriate tokens


value1 = int(input("Enter first number: "))
value2 = int(input("Enter second number: "))
operator = input("Enter operation to perform ( +,-,/,*)")

match operator :
  case '+':
    print(value1+value2)
  case '-':
    print(value1-value2)
  case '/':
    if(value2 == 0):
      print("Can not divide by 0.")
      exit
    print(value1/value2)
  case '*':
    print(value1 * value2)
  case _:
    print("Wrong steps : better luck next time ")
```
