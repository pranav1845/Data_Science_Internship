## `Control Flow` :

 Three main types of control flow :

 1. Sequential   (Default)
 2. Selection    (Decision)
 3. Repetation   (Loops)


 **Sequential** : In this the set of Instructions are executed in sequential manner (one after other). Computers by default follow this control flow.
