```python
#tuple of students and using (for) loop to iterate through it

classData = ("pranv", "pratik", "ganesh", "Sddhesh")

for s in classData:
  if s == "ganesh":
    print("He is from pune")
# marks iteration

marks = (
45, 60, 35, 75, 7,
12, 18, 23, 29, 56,
34, 38, 41, 45, 49,
53, 57, 60, 64, 68,
71, 75, 79, 83, 86,
89, 92, 95, 97, 100)
no_of_fail = 0
avg_fail_mark = 0

for mark in marks:
  if mark < 45:
    avg_fail_mark = avg_fail_mark + mark # suming marks of failed stduent to obtain avg fail mark
    no_of_fail = no_of_fail + 1
    print(f"Student {no_of_fail} : fail")
print("Total number of fail studnets ", no_of_fail)
print("worst marks in tuple", min(marks))
print("Average failing mark :", avg_fail_mark / no_of_fail)



```

```python
# Set in python

# Set items are unordered, unchangable, and do not allow duplicate values
# Unordered means they don't have predefined order
# Set does not allow duplicates values


sdata = {"apple", "pineapple", "orange", "pomogranade", "kiwi"}
# print Set
print(sdata)

# returns the type of data type
print(type(sdata))

#check if item is in set or not
print("orange" in sdata)

#updaing existing Set
new_set = {"mango", "cherry", "apple"}

sdata.update(new_set)

print(sdata)

# add() and remove() method of Set

sdata.add("Pinakal")

sdata.remove("orange")

# pop() method

sdata.pop()


```
